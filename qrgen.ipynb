```
import matplotlib.pyplot as plt
from IPython.display import Image, display
```

```
import qrcode
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data('Lorem ipsum dolor sit amet')
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="#aaaaaa")
display(img)
```
